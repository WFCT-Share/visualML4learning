```python
import matplotlib.pyplot as plt

def create_plot():
    fig, ax = plt.subplots()
    ax.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
    return fig, ax

def save_plot(fig, ax):
    ax.set_title("Example Plot")
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    fig.savefig("saved_plot.png")

fig, ax = create_plot()
save_plot(fig, ax)
```
